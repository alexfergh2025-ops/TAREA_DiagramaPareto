```python
#IMPORTAR LIBRERIAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

```

```python
#Cargar datos
uploaded = files.upload()

#Crear Dataframe
df = pd.read_csv(list(uploaded.keys())[0])

#Mostrar Dataframe
df
```

```python
#Ordenar los datos de mayor a menor
df = df.sort_values(by="Frecuencia", ascending=False)
#Crear nueva columna
df["Porcentaje"]=df["Frecuencia"] / df["Frecuencia"].sum()*100
df["Porcentaje Acumulado"]=df["Porcentaje"].cumsum()
df
```

```python
#Graficar garfica de pareto
fig, ax1 = plt.subplots(figsize=(10, 6))

#Grafico de barras
ax1.bar(df["Causa"], df["Frecuencia"], color="blue")
ax1.set_xticklabels(df["Causa"], rotation = 45, ha = 'right')
ax1.set_ylabel("Frecuencia", fontsize=12)

#Linea de porcentaje acumulado
ax2 = ax1.twinx()
ax2.plot(df["Causa"], df["Porcentaje Acumulado"], color="red", marker="o", linewidth=2)
ax2.set_ylabel("Porcentaje Acumulado (%)", fontsize=12)
ax2.set_ylim(0,120)

#Linea guia del 80%
ax2.axhline(y=80, color='red', linestyle='--',linewidth=1)

plt.title("Gráfica de Pareto - Clinica ABC Salud", fontsize=14)
plt.tight_layout()
plt.show()
```
